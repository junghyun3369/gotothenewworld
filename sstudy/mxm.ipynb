```python
# 1. 딕셔너리 생성하기
my_dict = {
    '이름': '홍길동',
    '나이': 25,
    '직업': '프로그래머'
}
print(f"초기 딕셔너리: {my_dict}")

# 2. '이름'에 해당하는 값 찾아 출력하기
print(f"이름: {my_dict['이름']}")

# 3. '사는곳': '서울' 키-값 쌍 추가하기
my_dict['사는곳'] = '서울'
print(f"'사는곳' 추가 후: {my_dict}")

# 4. '나이' 값 변경하기
my_dict['나이'] = 30  # '나이'를 30으로 변경
print(f"'나이' 변경 후: {my_dict}")

# 5. '직업' 키-값 쌍 삭제하기
del my_dict['직업']
print(f"'직업' 삭제 후 최종 딕셔너리: {my_dict}")
```

```python
def 전화번호부():
    전화번호부 = {}  # 빈 딕셔너리 생성


    while True:
        이름 = input("이름을 입력하세요 (종료하려면 '종료' 입력): ") # 사용자에게 이름을 입력받기


        if 이름 == '종료':
            print("프로그램을 종료합니다.") # '종료' 입력 시 프로그램 종료
            break


        if 이름 in 전화번호부:
            print(f"{이름}의 전화번호: {전화번호부[이름]}") # 전화번호부에서 이름에 해당하는 전화번호가 있는지 확인
        else:
            전화번호 = input(f"{이름}의 전화번호를 입력하세요: ") # 전화번호 추가
            전화번호부[이름] = 전화번호
            print(f"{이름}의 전화번호가 추가되었습니다.")


전화번호부() # 전화번호부 프로그램 실행
```

```python
요일 = ("월", "화", "수", "목", "금")
간식 = ["초코칩", "바나나", "요거트", "귤", "쿠키"]
# 요일은 튜플. 변경이 필요 없는 고정된 데이터에 적합
# 간식은 리스트. 변경 가능한 자료형이라 필요 시 수정이 쉬움
# 두 자료는 같은 순서로 나열돼 있어서 짝지을 수 있습니다.


추천표 = {}
# 비어 있는 딕셔너리를 만듭니다.
# 이 딕셔너리에 "요일:간식" 형태로 정보를 넣을 거예요.


for i in range(len(요일)):
    # 요일의 개수(5개)만큼 반복합니다. 즉, i는 0~4까지 돌아요.
    추천표[요일[i]] = 간식[i]
    # 예: 추천표["월"] = "초코칩" 처럼 딕셔너리에 값이 들어갑니다.


오늘 = input("오늘 요일을 입력하세요(월~금): ")
# 사용자가 오늘 요일을 직접 입력합니다.
# 입력된 값은 문자열 형태로 오늘 변수에 저장됩니다.


if 오늘 in 추천표:
    # 사용자가 입력한 요일이 딕셔너리에 있는지 확인합니다.
    print(f"오늘의 추천 간식은 {추천표[오늘]}입니다!")
    # 추천표[오늘]로 해당 요일의 간식을 가져와 출력합니다.
else: #없다면
    print("올바른 요일을 입력해주세요.")
    # "올바른 요일을 입력해주세요."라는 안내 문구를 보여줍니다.
```
